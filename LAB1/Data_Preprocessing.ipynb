#Feature encoding

```python
#Nominal and Binary Encoding
df = pd.get_dummies(df, columns=['gender'])
df.head()
```

```python
#change back to attribute gender with value 'M', 'F'
df['gender'] = df[['gender_M', 'gender_F']].idxmax(axis=1).str.replace('gender_', '')
#value in gender are gender_M, gender_F and replace('gender_', '') to get M, F and remove gender_M and gender_F attributes
df = df.drop(['gender_M', 'gender_F'], axis=1)
df
```

```python
#Ordinal Encoding
mapping = {'silver': 1,'gold': 2,'platinum': 3}
df['grade'] = df['grade'].map(mapping)
df.head()
```

#Apply Function

```python
#By defining a function beforehand.
# Apply function
def fun(value):
	if value > 70:
		return "Yes"
	else:
		return "No"

df['Customer Satisfaction'] = df['score'].apply(fun)
df.head(10)
```

```python
#By using the lambda operator to normalize age values between 0 and 1, 1 represents the oldest person
const = df['age'].max()
df['age'] = df['age'].apply(lambda x: x/const)
df.head()
```
